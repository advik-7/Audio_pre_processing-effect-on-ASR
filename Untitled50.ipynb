```python
from transformers import WhisperProcessor, WhisperForConditionalGeneration
from datasets import Audio
from pydub import AudioSegment
import numpy as np
import torch

# Load the Whisper model and processor
processor = WhisperProcessor.from_pretrained("openai/whisper-medium")
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-medium")

# Set language and task for forced decoding
forced_decoder_ids = processor.get_decoder_prompt_ids(language="hindi", task="transcribe")

# Load the audio file
audio_path = "/content/SandalWoodNewsStories_146.mp3"
audio = AudioSegment.from_mp3(audio_path)

# Convert audio to 16kHz mono channel (Whisper expects 16kHz audio)
audio = audio.set_frame_rate(16000).set_channels(1)

# Export the audio as a numpy array
audio_array = np.array(audio.get_array_of_samples())

# Prepare the audio input for Whisper
input_features = processor(audio_array, sampling_rate=16000, return_tensors="pt").input_features

# Generate transcription
predicted_ids = model.generate(input_features, forced_decoder_ids=forced_decoder_ids)

# Decode token IDs to text
transcription = processor.batch_decode(predicted_ids, skip_special_tokens=True)

# Output the transcription
print("Transcription:", transcription[0])
```

```python
import librosa
import soundfile as sf
import noisereduce as nr
import numpy as np
import scipy.signal as sps

# Load the audio file
input_path = "/content/SandalWoodNewsStories_146.wav"
output_path = "/content/processed_audio.wav"

# Step 1: Load and resample the audio to 16 kHz mono
audio, sr = librosa.load(input_path, sr=None, mono=True)
target_sr = 16000  # Standard ASR sampling rate
if sr != target_sr:
    audio = sps.resample(audio, int(len(audio) * target_sr / sr))
    sr = target_sr

# Step 2: Noise reduction (assuming background noise)
reduced_noise_audio = nr.reduce_noise(y=audio, sr=sr)

# Step 3: Silence removal
non_silent_intervals = librosa.effects.split(reduced_noise_audio, top_db=20)
processed_audio = np.concatenate([reduced_noise_audio[start:end] for start, end in non_silent_intervals])

# Save the processed audio
sf.write(output_path, processed_audio, sr)

# Step 4: Optional feature extraction (e.g., MFCC)
mfcc_features = librosa.feature.mfcc(y=processed_audio, sr=sr, n_mfcc=13)
print("MFCC features shape:", mfcc_features.shape)
```

```python
import matplotlib.pyplot as plt
import librosa.display

# Load the original and processed audio
original_audio, original_sr = librosa.load("/content/SandalWoodNewsStories_146.wav", sr=None, mono=True)
processed_audio, processed_sr = librosa.load("/content/processed_audio.wav", sr=None, mono=True)

# Plotting function for waveform
def plot_waveforms(original, processed, sr):
    plt.figure(figsize=(15, 6))

    plt.subplot(2, 1, 1)
    librosa.display.waveshow(original, sr=sr, color='blue')
    plt.title("Original Audio Waveform")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")

    plt.subplot(2, 1, 2)
    librosa.display.waveshow(processed, sr=sr, color='green')
    plt.title("Processed Audio Waveform")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")

    plt.tight_layout()
    plt.show()

# Plotting function for spectrogram
def plot_spectrograms(original, processed, sr):
    plt.figure(figsize=(15, 6))

    plt.subplot(2, 1, 1)
    original_spec = librosa.amplitude_to_db(np.abs(librosa.stft(original)), ref=np.max)
    librosa.display.specshow(original_spec, sr=sr, x_axis='time', y_axis='log', cmap='cool')
    plt.colorbar(format='%+2.0f dB')
    plt.title("Original Audio Spectrogram")

    plt.subplot(2, 1, 2)
    processed_spec = librosa.amplitude_to_db(np.abs(librosa.stft(processed)), ref=np.max)
    librosa.display.specshow(processed_spec, sr=sr, x_axis='time', y_axis='log', cmap='cool')
    plt.colorbar(format='%+2.0f dB')
    plt.title("Processed Audio Spectrogram")

    plt.tight_layout()
    plt.show()

# Plotting function for MFCCs
def plot_mfccs(original, processed, sr):
    plt.figure(figsize=(15, 6))

    plt.subplot(2, 1, 1)
    original_mfccs = librosa.feature.mfcc(y=original, sr=sr, n_mfcc=13)
    librosa.display.specshow(original_mfccs, x_axis='time', sr=sr)
    plt.colorbar()
    plt.title("Original Audio MFCCs")

    plt.subplot(2, 1, 2)
    processed_mfccs = librosa.feature.mfcc(y=processed, sr=sr, n_mfcc=13)
    librosa.display.specshow(processed_mfccs, x_axis='time', sr=sr)
    plt.colorbar()
    plt.title("Processed Audio MFCCs")

    plt.tight_layout()
    plt.show()

# Run the plotting functions
plot_waveforms(original_audio, processed_audio, original_sr)
plot_spectrograms(original_audio, processed_audio, original_sr)
plot_mfccs(original_audio, processed_audio, original_sr)
```

```python
import numpy as np
import librosa
import soundfile as sf
import matplotlib.pyplot as plt
import librosa.display

# Load the original audio
input_path = "/content/SandalWoodNewsStories_146.wav"
noisy_output_path = "/content/noisy_audio.wav"

audio, sr = librosa.load(input_path, sr=None, mono=True)

# Step 1: Add noise
# We will add white noise with a small amplitude
noise_factor = 0.05  # Increase this value for stronger noise
noisy_audio = audio + noise_factor * np.random.normal(0, 1, len(audio))

# Save the noisy audio file
sf.write(noisy_output_path, noisy_audio, sr)

# Plotting function for waveform
def plot_waveforms(original, noisy, sr):
    plt.figure(figsize=(15, 6))

    plt.subplot(2, 1, 1)
    librosa.display.waveshow(original, sr=sr, color='blue')
    plt.title("Original Audio Waveform")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")

    plt.subplot(2, 1, 2)
    librosa.display.waveshow(noisy, sr=sr, color='red')
    plt.title("Noisy Audio Waveform")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")

    plt.tight_layout()
    plt.show()

# Plotting function for spectrogram
def plot_spectrograms(original, noisy, sr):
    plt.figure(figsize=(15, 6))

    plt.subplot(2, 1, 1)
    original_spec = librosa.amplitude_to_db(np.abs(librosa.stft(original)), ref=np.max)
    librosa.display.specshow(original_spec, sr=sr, x_axis='time', y_axis='log', cmap='cool')
    plt.colorbar(format='%+2.0f dB')
    plt.title("Original Audio Spectrogram")

    plt.subplot(2, 1, 2)
    noisy_spec = librosa.amplitude_to_db(np.abs(librosa.stft(noisy)), ref=np.max)
    librosa.display.specshow(noisy_spec, sr=sr, x_axis='time', y_axis='log', cmap='cool')
    plt.colorbar(format='%+2.0f dB')
    plt.title("Noisy Audio Spectrogram")

    plt.tight_layout()
    plt.show()

# Plotting function for MFCCs
def plot_mfccs(original, noisy, sr):
    plt.figure(figsize=(15, 6))

    plt.subplot(2, 1, 1)
    original_mfccs = librosa.feature.mfcc(y=original, sr=sr, n_mfcc=13)
    librosa.display.specshow(original_mfccs, x_axis='time', sr=sr)
    plt.colorbar()
    plt.title("Original Audio MFCCs")

    plt.subplot(2, 1, 2)
    noisy_mfccs = librosa.feature.mfcc(y=noisy, sr=sr, n_mfcc=13)
    librosa.display.specshow(noisy_mfccs, x_axis='time', sr=sr)
    plt.colorbar()
    plt.title("Noisy Audio MFCCs")

    plt.tight_layout()
    plt.show()

# Run the plotting functions
plot_waveforms(audio, noisy_audio, sr)
plot_spectrograms(audio, noisy_audio, sr)
plot_mfccs(audio, noisy_audio, sr)
```
